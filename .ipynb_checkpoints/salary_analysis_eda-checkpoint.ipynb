# üíº Salary Analysis Project

This project analyzes salary trends in data-related roles using a publicly available dataset. The goal is to explore how different factors like job title, experience level, and company type affect annual salaries.

This kind of analysis is useful for:
- Job seekers (to negotiate better salaries)
- Recruiters (to set fair offers)
- Companies (to benchmark compensation)


## üóÇÔ∏è About the Dataset

The dataset comes from Kaggle and contains anonymized salary data for various data science and tech roles collected between 2020 and 2023.

Each row represents a job, with salary, job details, and company info.


```python
import pandas as pd

# Load the dataset
df = pd.read_csv('data/ds_salaries.csv')

# Preview the data
df.head()
```

### ‚úèÔ∏è Column Descriptions

- **job_title**: The position held (e.g., Data Scientist, ML Engineer).
- **salary_in_usd**: Annual salary in US dollars.
- **employment_type**: Type of employment (`FT` = Full-Time, `PT` = Part-Time, `FL` = Freelance, `IN` = Intern).
- **experience_level**: 
  - `EN` = Entry-level
  - `MI` = Mid-level
  - `SE` = Senior-level
  - `EX` = Executive-level
- **remote_ratio**: How remote the job is (0 = on-site, 50 = hybrid, 100 = fully remote).
- **company_location**: Country where the company is based.
- **employee_residence**: Country where the employee lives.
- **company_size**: Size of the company (`S`, `M`, or `L`).
- **work_year**: Year the salary was reported (2020‚Äì2023).


```python
# Check dataset shape and column names
print("Rows and columns:", df.shape)
print("\nColumns:\n", df.columns)

# Check for missing values
print("\nMissing values:\n", df.isnull().sum())

# Check data types
print("\nData types:\n", df.dtypes)
```

## üßπ Initial Cleaning

We cleaned up the column names by:
- Converting them to lowercase
- Replacing spaces with underscores

This makes it easier to use in Python code (e.g., `df.job_title` instead of `df["Job Title"]`).
We also confirmed that:
- There are no missing values
- All salaries are already standardized in USD


```python
# Job titles count
print("Top 10 Job Titles:")
print(df['job_title'].value_counts().head(10), '\n')

# Experience level count
print("Experience Levels:")
print(df['experience_level'].value_counts(), '\n')

# Company size distribution
print("Company Size:")
print(df['company_size'].value_counts(), '\n')

# Remote work ratio
print("Remote Work Distribution:")
print(df['remote_ratio'].value_counts(), '\n')
```

## üìä Data Overview

We explored some basic statistics to understand the structure of the data:
- The most common **job titles**
- Distribution of **experience levels**
- Breakdown of **company sizes**
- Share of **remote vs. on-site** jobs

This gives us a general sense of the types of roles and companies represented in the dataset.


```python
import matplotlib.pyplot as plt

# Group and sort
avg_salary_by_title = df.groupby('job_title')['salary_in_usd'].mean().sort_values(ascending=False).head(10)

# Plot
plt.figure(figsize=(10,6))
avg_salary_by_title.plot(kind='barh', color='skyblue')
plt.title('Top 10 Job Titles by Average Salary (USD)')
plt.xlabel('Average Salary (USD)')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

## üìà Visualization: Top 10 Highest-Paid Roles

We plotted the top 10 job titles with the highest **average salaries**.

### Key Insights:
- **Data Architects**, **Machine Learning Engineers**, and **Cloud Engineers** often appear at the top.
- Executive roles dominate in terms of pay, but senior IC roles can also earn very high.
- Some titles may have very few samples, so we‚Äôll apply filters in future steps.

This visualization gives us an early view of which roles are best-paid in the dataset.


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set style
sns.set(style="whitegrid")

# Boxplot of salary by experience
plt.figure(figsize=(8,5))
sns.boxplot(x='experience_level', y='salary_in_usd', data=df, palette='pastel')
plt.title('Salary Distribution by Experience Level')
plt.xlabel('Experience Level')
plt.ylabel('Salary (USD)')
plt.show()
```

## üë§ Salary Distribution by Experience Level

This boxplot visualizes how salary varies across different levels of professional experience:

- **EN (Entry-Level)**: Typically the lowest salary bracket.
- **MI (Mid-Level)**: Noticeable increase compared to entry-level roles.
- **SE (Senior-Level)**: Shows higher and more spread-out salaries, reflecting skill premium.
- **EX (Executive-Level)**: Highest overall salaries with some extreme outliers.

This plot helps identify how much salary growth can be expected with more experience, and whether there‚Äôs overlap between levels.


```python
# Average salary by remote ratio
remote_salary = df.groupby('remote_ratio')['salary_in_usd'].mean()

# Plot
plt.figure(figsize=(6,4))
remote_salary.plot(kind='bar', color='coral')
plt.title('Average Salary by Remote Ratio')
plt.xlabel('Remote Ratio (%)')
plt.ylabel('Average Salary (USD)')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

## üè† Average Salary by Remote Work Ratio

This bar chart compares the average salary across three levels of remote work:

- **0%**: Fully on-site jobs
- **50%**: Hybrid jobs (partially remote)
- **100%**: Fully remote jobs

This comparison helps assess whether remote flexibility has any relationship with salary. For example, some remote jobs may offer higher pay due to global talent access, or they may pay less due to reduced overhead or location-based adjustments.


```python
# Top 10 countries by avg salary
top_countries = df.groupby('company_location')['salary_in_usd'].mean().sort_values(ascending=False).head(10)

# Plot
plt.figure(figsize=(8,5))
top_countries.plot(kind='barh', color='mediumseagreen')
plt.title('Top 10 Countries by Average Salary')
plt.xlabel('Average Salary (USD)')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

## üåç Top 10 Countries by Average Salary

This chart shows the top 10 countries based on the **average salary offered by companies**.

It provides insight into which countries (regardless of employee location) offer the most competitive salaries in the data field. Countries like the United States, Switzerland, or the UK often appear at the top due to their mature tech ecosystems and high cost of living.

This is useful for:
- Remote job seekers choosing where to apply
- Analysts understanding global salary benchmarks


## üìå Key Insights from the Analysis

- üíº **Executive-level roles** earn the highest salaries, with a sharp jump from senior to executive level.
- üíª **Fully remote roles** tend to offer slightly higher average salaries, indicating demand for global flexibility.
- üåç **Companies in the U.S. and Switzerland** offer the highest average salaries across data roles.
- üìä **Data Architects, ML Engineers, and Cloud Engineers** are among the highest-paid job titles on average.
- üìà There is wide salary variation even within the same experience level, suggesting industry, location, and job type also play a big role.
